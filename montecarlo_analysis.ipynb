```python
#########################
### IMPORT STATEMENTS ###
#########################

### Standard libraries

import pandas as pd
import numpy as np
import os

import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Path to the directory containing the Excel files
directory_path = './Data/Results'

# Initialize an empty list to store dataframes
dfs = []

# Iterate over each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.xlsx'):
        # Construct the full file path
        file_path = os.path.join(directory_path, filename)
        
        # Read the Excel file into a dataframe
        df = pd.read_excel(file_path)
        
        # Append the dataframe to the list
        dfs.append(df)

# Concatenate all dataframes in the list along the rows
df_all = pd.concat(dfs, ignore_index=True)

# Define a custom function to convert strings to numeric values
def convert_to_numeric(value):
    try:
        # Extract numeric part from the string and convert to float
        numeric_value = float(value.replace('[', '').replace(']', ''))
        return numeric_value
    except ValueError:
        # Return NaN if conversion fails
        return float('nan')

# Convert the 'Chi_error' column to numeric values using the custom function
df_all['Chi_error'] = df_all['Chi_error'].apply(convert_to_numeric)
df_all['Chi_error'] = abs(df_all['Chi_error'])

df_all
```

```python
df_avg = df_all.groupby(['Century', 'Type']).agg({'Ox Range Size': 'mean', 
                                                  'Chi_error': 'mean', 
                                                  'Chi_range_size': 'mean'}).reset_index()


custom_order_type = ['SA', 'SB','SQ', 'MA', 'MB', 'MQ', 'LA', 'LB', 'LQ']
custom_order_century = sorted(df_avg['Century'].unique(), key=lambda x: int(x[:-2]), reverse=True)


df_avg = df_avg.assign(sort_key_century=df_avg['Century'].map(lambda x: custom_order_century.index(x)))
df_avg = df_avg.assign(sort_key_type=df_avg['Type'].map(lambda x: custom_order_type.index(x)))
df_avg['sort_key'] = df_avg['sort_key_century'] * len(custom_order_type) + df_avg['sort_key_type']
df_avg = df_avg.sort_values(by='sort_key').drop(columns=['sort_key', 'sort_key_century', 'sort_key_type'])

print(df_avg)
```

```python
######## PREPARE DATA FOR HEATMAPS

df_heat_oxcal = df_avg.explode('Ox Range Size')
heatmap_data_oxcal = df_heat_oxcal.pivot_table(index='Century', columns='Type', values='Ox Range Size', aggfunc='mean')
df_heat_chirange = df_avg.explode('Chi_range_size')
heatmap_data_chirange = df_heat_chirange.pivot_table(index='Century', columns='Type', values='Chi_range_size', aggfunc='mean')

custom_order_x = ['SA', 'SB','SQ', 'MA', 'MB', 'MQ', 'LA', 'LB', 'LQ']
custom_order_y = sorted(df['Century'].unique(), key=lambda x: int(x[:-2]), reverse=True)

heatmap_data_oxcal = heatmap_data_oxcal[custom_order_x]
heatmap_data_oxcal = heatmap_data_oxcal.reindex(custom_order_y)

heatmap_data_chirange = heatmap_data_chirange[custom_order_x]
heatmap_data_chirange = heatmap_data_chirange.reindex(custom_order_y)
```

```python
########### FIGURE 2

# Set up the figure and axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 15))

# Plot the first heatmap
sns.heatmap(data=heatmap_data_oxcal, cmap='plasma_r', ax=ax1, cbar=False,
            xticklabels=custom_order_x, yticklabels=custom_order_y, alpha=0.9)

# Apply custom order to y axis
ax1.set_yticklabels(ax1.get_yticklabels(), fontsize=12, fontweight='bold', rotation=0)
ax1.set_xticklabels(ax1.get_xticklabels(), fontsize=12, fontweight='bold')

# Set labels and title for the first plot
ax1.set_xlabel('')
ax1.set_ylabel('')


# Plot the second heatmap
sns.heatmap(data=heatmap_data_chirange, cmap='plasma_r', ax=ax2, cbar=False,
            xticklabels=custom_order_x, yticklabels=custom_order_y, alpha=0.9)

# Apply custom order to y axis
ax2.set_yticklabels(ax2.get_yticklabels(), fontsize=12, fontweight='bold', rotation=0)
ax2.set_xticklabels(ax2.get_xticklabels(), fontsize=12, fontweight='bold')

# Set labels and title for the second plot
ax2.set_xlabel('')
ax2.set_ylabel('')


# Create a shared colorbar
cbar = fig.colorbar(ax1.collections[0], ax=[ax1, ax2], orientation='vertical')
cbar.set_label('', fontsize=12, fontweight='bold')

fig.text(0.45, 0.05, 'Simulation Type', ha='center', fontsize=20, fontweight='bold')
fig.text(0.08, 0.5, 'Century [CE]', va='center', rotation='vertical', fontsize=20, fontweight='bold')
fig.text(0.84, 0.5, 'Range [years]', va='center', rotation=-90, fontsize=20, fontweight='bold')
fig.text(0.265, 0.90, 'OxCal', ha='center', fontsize=20, fontweight='bold')
fig.text(0.605, 0.90, '$\chi^2$', ha='center', fontsize=20, fontweight='bold')

# Save and show the plot
plt.savefig(f'./Figures/Overview/Montecarlo/heatmap_comparison.jpg', bbox_inches='tight', pad_inches=0.5, dpi=300)
plt.show()
```
